```python
import random
import math
import numpy as np

class GA2048:
    def __init__(self, gridSize, populationSize):
        self.population = dict()
        self.gridSize = gridSize
        self.populationSize = populationSize
        self.totalFitness = 0

        i = 0
        while i < populationSize:
            individual = np.random.randint(0, 15, size=(gridSize, gridSize))
            self.population[i] = [individual, self.calculateFitness(individual)]
            self.totalFitness += self.population[i][1]
            i += 1

    def calculateFitness(self, individual):
        # Sum up the values in the individual matrix as a simple fitness measure
        return np.sum(individual)


    def updatePopulationFitness(self):
      self.totalFitness = sum(self.population[individual][1] for individual in self.population)


    def selectParents(self):
      rouletteWheel = []
      wheelSize = self.populationSize * 5
      h_n = [self.population[individual][1] for individual in self.population]

      for j in range(self.populationSize):
          individualLength = round(wheelSize * (h_n[j] / sum(h_n)))
          if individualLength > 0:
              rouletteWheel.extend([j] * individualLength)

      random.shuffle(rouletteWheel)
      parentIndices = [rouletteWheel[random.randint(0, len(rouletteWheel) - 1)] for _ in range(self.populationSize)]
      newGeneration = {i: self.population[parentIndices[i]].copy() for i in range(self.populationSize)}
      del self.population
      self.population = newGeneration.copy()
      self.updatePopulationFitness()


    def generateChildren(self, crossoverProbability):
        numberOfPairs = round(crossoverProbability * self.populationSize / 2)
        individualIndices = list(range(self.populationSize))
        random.shuffle(individualIndices)

        i, j = 0, 0
        while i < numberOfPairs:
            crossoverPoint = random.randint(0, self.gridSize - 1)
            child1 = np.vstack((self.population[j][0][:crossoverPoint], self.population[j+1][0][crossoverPoint:]))
            child2 = np.vstack((self.population[j+1][0][:crossoverPoint], self.population[j][0][crossoverPoint:]))

            self.population[j] = [child1, self.calculateFitness(child1)]
            self.population[j+1] = [child2, self.calculateFitness(child2)]

            i += 1
            j += 2

    def mutateChildren(self, mutationProbability):
      numberOfBits = round(mutationProbability * self.populationSize * self.gridSize * self.gridSize)
      totalIndices = list(range(self.populationSize * self.gridSize * self.gridSize))
      random.shuffle(totalIndices)
      swapLocations = random.sample(totalIndices, numberOfBits)

      for loc in swapLocations:
          individualIndex, remainder = divmod(loc, self.gridSize * self.gridSize)
          row, col = divmod(remainder, self.gridSize)
          self.population[individualIndex][0][row, col] = 1 - self.population[individualIndex][0][row, col]

      self.updatePopulationFitness()


# Example:
gridSize, populationSize = 4, 10
i = 0
instance = GA2048(gridSize, populationSize)
# while i<=100:
while True:
    instance.selectParents()
    instance.generateChildren(0.8)
    instance.mutateChildren(0.03)

    print(instance.population)
    print(instance.totalFitness)
    print(i)

    i += 1
    found = any(instance.population[individual][1] == gridSize for individual in instance.population)

    if found:
        break
```
